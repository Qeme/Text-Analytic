```python
import scrapy
import re
from scrapy.crawler import CrawlerProcess

#create a class for spider
class MySpider(scrapy.Spider):
    #name of the spider
    name = 'myspider'
    #to make sure the spider will stick to one page only
    allowed_domains = ['amazon.com']
    #the link to start the scrapy spider deployment
    start_urls = ['https://www.amazon.com/Stanley-IceFlow-Stainless-Steel-Tumbler/product-reviews/B0CT4BB651/ref=cm_cr_getr_d_paging_btm_prev_1?ie=UTF8&reviewerType=all_reviews&pageNumber=1']
    page_number = 1
    
    def parse(self,response):
        #do loop for all books (<article class="product_pod"></article>)
        for container in response.xpath('//div[@class="a-section review aok-relative"]'):
            #under article there are images (div class="image_container/a tag/img tag/the src of the image")
            username = container.xpath('.//div[@class="a-profile-content"]/span/text()').get()
            #under article there are price (div class="product_price")+text() is necessary to just take the string value not the whole syntax
            #\w+ matches one or more word characters (letters, digits, or underscores) for the month.
            #\s+ matches one or more whitespace characters.
            #\d{1,2} matches one or two digits for the day.
            #, matches a comma.
            #\d{4} matches exactly four digits for the year.
            date_review = container.xpath('.//span[@class="a-size-base a-color-secondary review-date"]/text()').get('').strip()
            date_extract = re.findall (r'on (\w+\s+\d{1,2},\s+\d{4})',date_review)
            # getall() is a function is to take all the string
            review_text = container.xpath('.//div[@class="a-row a-spacing-small review-data"]/span[@class="a-size-base review-text review-text-content"]/span/text()').getall()
            yield {'username': username,
                  'date_review': date_extract,
                  'review_text': review_text
                  }
        
        # next page go copy the link for page 
        next_page = "https://www.amazon.com/Stanley-IceFlow-Stainless-Steel-Tumbler/product-reviews/B0CT4BB651/ref=cm_cr_getr_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&"+ str(MySpider.page_number) +"/"
        if MySpider.page_number <= 5:
            MySpider.page_number +=1
            
            yield response.follow(next_page, callback=self.parse)

# give output inside a csv file in csv format
process = CrawlerProcess(settings = {'FEEDS': {'OutputAmazon.csv':{'format':'csv'}}})
#now run the code to crawl inside the webpage
process.crawl(MySpider)
process.start()
```
